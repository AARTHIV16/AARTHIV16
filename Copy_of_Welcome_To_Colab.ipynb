```
# Box Dimension Estimation from User-Provided Image using MiDaS

# ----------------------
!pip install -q opencv-python-headless torch matplotlib Pillow
!apt-get install -y libgl1-mesa-glx
#-----------------------
# 0. Install Dependencies
# ---
# ------------------------
# 1. Import Required Libraries
# ------------------------
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import contextlib
from google.colab import files

# Set matplotlib backend for Colab compatibility
%matplotlib inline

# ------------------------
# 2. Image Upload & Handling
# ------------------------
uploaded = files.upload()
image_path = next(iter(uploaded))
img = Image.open(image_path).convert("RGB")

# Resize image to reduce processing load
resize_factor = 2
img = img.resize((img.width//resize_factor, img.height//resize_factor))
img_np = np.array(img)
img_cv = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)

# Display the image first
plt.figure()
plt.imshow(img)
plt.title("Identify 3 box corners (order: length, width, height)")
plt.show()

# Manually input coordinates
clicked_points = []
for i in range(3):
    x = int(input(f"Enter x-coordinate for point {i+1}: "))
    y = int(input(f"Enter y-coordinate for point {i+1}: "))
    clicked_points.append((x, y))
    print(f"Point {i+1}: ({x}, {y})")

# ------------------------
# 3. Load MiDaS Model (with suppressed output)
# ------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_type = "DPT_Large"

with open(os.devnull, 'w') as f, contextlib.redirect_stdout(f):
    midas = torch.hub.load('intel-isl/MiDaS', model_type).to(device)
    midas.eval()
    midas_transforms = torch.hub.load('intel-isl/MiDaS', 'transforms')
    transform = midas_transforms.dpt_transform

# ------------------------
# 4. Depth Map Estimation
# ------------------------
input_batch = transform(img_np).to(device)

with torch.no_grad():
    prediction = midas(input_batch)
    prediction = torch.nn.functional.interpolate(
        prediction.unsqueeze(1),
        size=img_np.shape[:2],
        mode="bicubic",
        align_corners=False
    ).squeeze()

depth_map = prediction.cpu().numpy()

# Display depth map
plt.figure()
plt.imshow(depth_map, cmap='inferno')
plt.title("Estimated Depth Map")
plt.colorbar()
plt.show()

# Print depths for clicked points
for x, y in clicked_points:
    print(f"Point ({x}, {y}), Depth: {depth_map[y,x]:.2f}")

# ------------------------
# 7. 3D Calculations
# ------------------------
def calculate_3d_coords(x, y):
    fx = fy = 500  # Focal length assumption
    cx, cy = img.width//2, img.height//2
    Z = depth_map[y, x]
    return [(x - cx) * Z/fx, (y - cy) * Z/fy, Z]

dimensions = []
for i in range(3):
    p1 = calculate_3d_coords(*clicked_points[i])
    p2 = calculate_3d_coords(*clicked_points[(i+1)%3])
    dimensions.append(np.linalg.norm(np.array(p1) - np.array(p2)))

print("\nEstimated Dimensions (relative units):")
print(f"Length: {dimensions[0]:.2f}")
print(f"Width: {dimensions[1]:.2f}")
print(f"Height: {dimensions[2]:.2f}")
```

```
# ------------------------
# 8. Scale to Real-World Dimensions
# ------------------------

# Ask user for two reference points on an object with known real-world size
print("\nPlease click two points on a known reference object (e.g., A4 paper, credit card)")
ref_points = []
for i in range(2):
    x = int(input(f"Enter x-coordinate for reference point {i+1}: "))
    y = int(input(f"Enter y-coordinate for reference point {i+1}: "))
    ref_points.append((x, y))
    print(f"Reference Point {i+1}: ({x}, {y})")

# Ask for real-world distance between those two points (e.g., 21 cm for A4 width)
real_distance_cm = float(input("\nEnter the real-world distance between those points (in cm): "))

# Convert the 2 reference points into 3D space
ref_p1 = calculate_3d_coords(*ref_points[0])
ref_p2 = calculate_3d_coords(*ref_points[1])

# Calculate distance in 3D space (same unit as earlier box dimensions)
ref_distance = np.linalg.norm(np.array(ref_p1) - np.array(ref_p2))

# Calculate scale factor: cm per 3D unit
scale_factor = real_distance_cm / ref_distance
print(f"\nScale factor: {scale_factor:.4f} cm per unit")

# Convert previously estimated dimensions to real-world scale
real_dimensions = [d * scale_factor for d in dimensions]

print("\n Estimated Real-World Box Dimensions:")
print(f"Length: {real_dimensions[0]:.2f} cm")
print(f"Width : {real_dimensions[1]:.2f} cm")
print(f"Height: {real_dimensions[2]:.2f} cm")
```

```
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import contextlib
from google.colab import files

%matplotlib inline

# 1. Upload image
print("Upload your image file:")
uploaded = files.upload()
image_path = next(iter(uploaded))
img = Image.open(image_path).convert("RGB")

# Resize image (optional)
resize_factor = 2
img = img.resize((img.width // resize_factor, img.height // resize_factor))
img_np = np.array(img)
img_cv = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)

# 2. Contour detection to find box corners
def find_box_corners(image_bgr):
    gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    edges = cv2.Canny(blur, 50, 150)

    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Find largest quadrilateral contour
    max_area = 0
    box_contour = None
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 1000:  # filter small contours
            peri = cv2.arcLength(cnt, True)
            approx = cv2.approxPolyDP(cnt, 0.02 * peri, True)
            if len(approx) == 4 and area > max_area:
                max_area = area
                box_contour = approx

    if box_contour is None:
        raise ValueError("No box contour detected.")

    # Reshape and order points: top-left, top-right, bottom-right, bottom-left
    pts = box_contour.reshape(4, 2)
    rect = order_points(pts)
    return rect.astype(int)

def order_points(pts):
    # Order points in consistent order: tl, tr, br, bl
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]  # top-left has smallest sum
    rect[2] = pts[np.argmax(s)]  # bottom-right has largest sum

    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]  # top-right has smallest diff
    rect[3] = pts[np.argmax(diff)]  # bottom-left has largest diff
    return rect

try:
    box_corners = find_box_corners(img_cv)
    print("Detected box corners (pixel coordinates):")
    for i, (x,y) in enumerate(box_corners):
        print(f"P{i+1}: ({x}, {y})")
except Exception as e:
    print("Error detecting box corners:", e)
    # Fallback: manual input
    box_corners = []
    for i in range(4):
        x = int(input(f"Enter x-coordinate for point {i+1}: "))
        y = int(input(f"Enter y-coordinate for point {i+1}: "))
        box_corners.append((x,y))
    box_corners = np.array(box_corners)

# 3. Load MiDaS and estimate depth
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_type = "DPT_Large"

with open(os.devnull, 'w') as f, contextlib.redirect_stdout(f):
    midas = torch.hub.load('intel-isl/MiDaS', model_type).to(device)
    midas.eval()
    midas_transforms = torch.hub.load('intel-isl/MiDaS', 'transforms')
    transform = midas_transforms.dpt_transform

input_batch = transform(img_np).to(device)

with torch.no_grad():
    prediction = midas(input_batch)
    prediction = torch.nn.functional.interpolate(
        prediction.unsqueeze(1),
        size=img_np.shape[:2],
        mode="bicubic",
        align_corners=False
    ).squeeze()

depth_map = prediction.cpu().numpy()

# Show depth map
plt.figure(figsize=(8,6))
plt.imshow(depth_map, cmap='inferno')
plt.title("Estimated Depth Map (Relative Depth)")
plt.colorbar(label='Depth (relative units)')
plt.axis('off')
plt.show()

# 4. Camera intrinsic matrix (example laptop webcam calibration)
camera_matrix = np.array([[920, 0, img.width/2],
                          [0, 920, img.height/2],
                          [0, 0, 1]])

print("\nUsing Camera Intrinsic Matrix:")
print(camera_matrix)

# 5. Back-project 4 corners to 3D points
def backproject_to_3d(x, y, depth, K):
    fx, fy = K[0,0], K[1,1]
    cx, cy = K[0,2], K[1,2]
    Z = depth
    X = (x - cx) * Z / fx
    Y = (y - cy) * Z / fy
    return np.array([X, Y, Z])

points_3d = []
for (x,y) in box_corners:
    d = depth_map[y, x]
    points_3d.append(backproject_to_3d(x, y, d, camera_matrix))
points_3d = np.array(points_3d)

# 6. Estimate box dimensions (length, width, height) from 3D points
# Edges: (P1-P2), (P2-P3), (P3-P4), (P4-P1)
edges = [np.linalg.norm(points_3d[i] - points_3d[(i+1)%4]) for i in range(4)]

# Assuming box edges: length and width are adjacent edges (P1-P2 and P2-P3)
length_rel = edges[0]
width_rel = edges[1]

# Height estimation is tricky with a single image and flat box corners.
# You can approximate height by depth difference of top and bottom edges or ask user input.
# Here, we estimate height as average depth difference between top and bottom edges:
height_rel = abs(np.mean(points_3d[[0,1],2]) - np.mean(points_3d[[2,3],2]))

print("\nEstimated Dimensions (relative units):")
print(f"Length: {length_rel:.4f}")
print(f"Width: {width_rel:.4f}")
print(f"Height (approx.): {height_rel:.4f}")

# 7. Normalize scale by average edge length ~ 1 meter
avg_edge = (length_rel + width_rel) / 2
scale_factor = 1 / avg_edge
print(f"\nNormalized scale factor: {scale_factor:.4f}")

length_m = length_rel * scale_factor
width_m = width_rel * scale_factor
height_m = height_rel * scale_factor

print("\nEstimated Box Dimensions (meters):")
print(f"Length: {length_m:.3f} m")
print(f"Width: {width_m:.3f} m")
print(f"Height: {height_m:.3f} m")

# Convert to centimeters
length_cm = length_m * 100
width_cm = width_m * 100
height_cm = height_m * 100

print("\nEstimated Box Dimensions (centimeters):")
print(f"Length: {length_cm:.1f} cm")
print(f"Width: {width_cm:.1f} cm")
print(f"Height: {height_cm:.1f} cm")

# 8. Visualize detected box and edges on image
annotated_img = img_np.copy()
annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_RGB2BGR)

colors = {
    'length': (0, 0, 255),  # Red
    'width': (0, 255, 0),   # Green
    'height': (255, 0, 0),  # Blue (used for height visualization if needed)
}

# Draw edges (length and width)
cv2.line(annotated_img, tuple(box_corners[0]), tuple(box_corners[1]), colors['length'], 3)
cv2.line(annotated_img, tuple(box_corners[1]), tuple(box_corners[2]), colors['width'], 3)
cv2.line(annotated_img, tuple(box_corners[2]), tuple(box_corners[3]), (0,255,255), 3)  # Other edges in yellow
cv2.line(annotated_img, tuple(box_corners[3]), tuple(box_corners[0]), (0,255,255), 3)

# Add legend
cv2.putText(annotated_img, 'Length (Red)', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, colors['length'], 2)
cv2.putText(annotated_img, 'Width (Green)', (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.8, colors['width'], 2)

annotated_img_rgb = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 8))
plt.imshow(annotated_img_rgb)
plt.title("Detected Box with Color-Coded Edges")
plt.axis('off')
plt.show()
```
